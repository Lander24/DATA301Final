```
import numpy as np 

 #Takes a string representing a draft position and returns the round of that pick in the draft
def get_draft_round(row):
    if isinstance(row["Drafted (tm/rnd/yr)"], float):
      return np.nan
    else:
      return row["Drafted (tm/rnd/yr)"].split("/")[1][1]

#Takes a string representing a draft position and returns the overall pick in the draft
def get_draft_pick(row):
    if isinstance(row["Drafted (tm/rnd/yr)"], float):
      return np.nan
    else:
      return row["Drafted (tm/rnd/yr)"].split(" ")[-4][0:-2]
       
```

This loads in 22 csvs representing 11 years of offense and deffensive combine players it then cleans up some cols while adding new ones before appending the dataframes to one master dataframe for exporting

```
import io
import pandas as pd
df_master = None
year =2009
for i in range(0, 22):
  prefix ="d"
  if(i%2 ==0):
    year = year+1
    prefix ="o"
  s = "https://raw.githubusercontent.com/Lander24/DATA301Final/master/CombData/" + prefix  +str(year) + ".csv"
  df_temp = pd.read_csv(s)
  #droping uneeded rank and college col
  df_temp = df_temp.drop(["Rk", "College"], axis=1)
  #cleaning up player names
  df_temp["Player"] = df_temp.apply(lambda row:  row["Player"].split('\\')[0],  axis=1)
  #Add a col for a players height in inches
  df_temp["Height_Inches"] =  df_temp.apply(lambda row:  int(row["Height"].split('-')[0])*12 + int(row["Height"].split('-')[1]),  axis=1)
  df_temp["Round"] = df_temp.apply(get_draft_round, axis=1)
  df_temp["Pick"] = df_temp.apply(get_draft_pick, axis=1)
  if(i==0):
    df_master =df_temp
  else:
    df_master = df_master.append(df_temp)
df_master.to_csv("master.csv")

from google.colab import files
files.download('master.csv')
```
