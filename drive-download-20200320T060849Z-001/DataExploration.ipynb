```
positions = df_master["Pos"].unique()
drills = ["Height_Inches"	,"Wt",	"40YD"	,"Vertical",	"BenchReps",	"Broad Jump"	,"3Cone",	"Shuttle"]
dict_ranges = {}
for pos in positions:
  pos_mask =df_master["Pos"] == pos
  df_pos = df_master[pos_mask]
  dict_ranges[pos] = {}
  print(df_pos["40YD"].max())
  for drill in drills:
    dict_ranges[pos][drill] ={"max" : df_pos[drill].max() , "min" : df_pos[drill].min()}

  print(dict_ranges[pos])
```

```
diffs =[]
for pos in positions:
  pos_mask =df_master["Pos"] == pos
  df_pos = df_master[pos_mask]
  dict_ranges[pos] = {}
  print(df_pos["40YD"].max())
  for drill in drills:
    dict_ranges[pos][drill] ={"max" : df_pos[drill].max() , "min" : df_pos[drill].min()}
    df_draft_undraft = (df_pos.groupby("Drafted"))
    if(drill == "Shuttle" or drill == "3Cone" or drill == "40YD"):
      diffs.append((-1*( df_draft_undraft[drill].mean()["Drafted"]-df_draft_undraft[drill].mean()["Undrafted"])/(dict_ranges[pos][drill]["max"] - dict_ranges[pos][drill]["min"]), pos, drill))
    else:
      diffs.append(((df_draft_undraft[drill].mean()["Drafted"]-df_draft_undraft[drill].mean()["Undrafted"])/(dict_ranges[pos][drill]["max"] - dict_ranges[pos][drill]["min"]), pos, drill))

    (df_draft_undraft[drill].mean().
    plot.bar(title=(pos + " " + drill)))
    plt.ylim(dict_ranges[pos][drill]["min"], dict_ranges[pos][drill]["max"])

    plt.show()

```
